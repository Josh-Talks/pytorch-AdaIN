```python
from AdaIN.Datasets import PatchPositionDataset, load_h5
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_array_(dataset):
    raws = np.zeros((len(dataset), dataset[0].shape[1], dataset[0].shape[2]), dtype=np.float32)
    for i, raw in enumerate(dataset):
        raws[i,:] = raw
    return raws
```

```python
patch_positions = load_h5("/scratch/talks/data/epfl/classification_patches/epfl_resized_01_80x80.h5", "training")
```

```python
print(patch_positions.shape)
```

```python
trial_dataset = PatchPositionDataset(
    raw_path = "/scratch/talks/data/pixel_resize/epfl/source_mitoEM_true.h5",
    raw_key = "train_raw",
    patch_starts = patch_positions,
    n_samples =10,
    random_seed = 42,
)
```

```python
print(len(trial_dataset))
```

```python
raw_imgs = get_array_(trial_dataset)
```
